```python
import pandas as pd
df = pd.read_csv("alabama.csv")

df.head()
```

```python
df = df.rename(columns={'Year': 'Year_month'})
df.head()
```

```python
df.columns
```

```python
df_filtered = df[(df['year'].between(1985, 1989)) | (df['year'].between(2011, 2024))]
soucet_za_rok = df.groupby('year')['Homicide_count'].sum().reset_index()
print(soucet_za_rok)
```

```python
soucet_za_rok.to_csv("alabama_withouspike.csv", sep=';', index=False)
```

```python
df_filtered = df[(df['year'].between(1990,2010))  &  (df['month']==(12))]
print (df_filtered)
```

```python
df_filtered.to_csv("alabama_spike.csv", sep=';', index=False)
```
