```python
import pandas as pd
df = pd.read_csv("Execution_1976_2025.csv")

df.head()
```

```python
df.columns
```

```python
df["Name"] = df[["First Name", "Middle Name(s)", "Last Name", "Suffix"]].fillna("").astype(str).agg(" ".join, axis=1)

df.head()
```

```python
df = df[["Name"] + [col for col in df.columns if col != "Name"]]
df.head()
```

```python
df.to_csv("Exectuion_join.csv", sep=';', index=False)
```
