```python
import pandas as pd
df = pd.read_csv("Population_Race.csv")
df.columns
```

```python
# odstranění .0 převedením na int
df['White Alone'] = df['White Alone'].astype(str).str.replace(r'\.0$', '', regex=True)
df.head()
```

```python
# seznam sloupců, které chceš převést
cols = ['Black Alone', 'American Indian or Alaskan Native', 'Asian Alone', 'Hawaiian or Pacific Islander Alone', 'Two or More Races', 'Not Hispanic', 'Hispanic']

# převod všech na celé číslo
for col in cols:
    df[col] = df[col].astype(str).str.replace(r'\.0$', '', regex=True)
df.head()
```

```python
for col in cols:
    df[col] = df[col].astype(str)
df.head()
```

```python
df.to_csv("Population_Rase_beznul.csv", sep=';', index=False)
```

```python
df.dtypes
```

```python
cols = ['White Alone','Black Alone', 'American Indian or Alaskan Native', 'Asian Alone', 'Hawaiian or Pacific Islander Alone', 'Two or More Races', 'Not Hispanic', 'Hispanic']
```

```python
for col in cols:
    df[col] = pd.to_numeric(df[col], errors='coerce').astype('Int64')
```

```python
df.to_csv("Population_Rase_beznul.csv", sep=';', index=False)
df.head()
```
