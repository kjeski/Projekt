```python
import pandas as pd
df = pd.read_csv("Population_part2_beznul.csv", sep=";", encoding="utf-8-sig")
df.head()
```

```python

# 6️⃣ "Rozbalení" sloupců od White Alone dál do řádků
# Prvních 5 sloupců zůstanou jako identifikátory (id_vars)
df_melted = df.melt(
    id_vars=df.columns[:3],     # prvních 5 sloupců zůstane
    var_name="Race",             # nový sloupec pro názvy ras
    value_name="Population"      # nový sloupec pro hodnoty
)

# 7️⃣ Ulož do nového souboru
df_melted.to_csv("Population_Race_long.csv", sep=";", index=False, encoding="utf-8-sig")


```
