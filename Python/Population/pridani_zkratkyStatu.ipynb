```python
import pandas as pd
df = pd.read_csv("Population_Race_long.csv", sep=";", encoding="utf-8-sig")
states = pd.read_csv("usa_state_abbreviations.csv",  encoding="utf-8-sig")


states.head()
```

```python
df.head()
```

Pridani sloupce se zkratkama

```python
main_path   = "Population_Race_long.csv"   # hlavní tabulka
states_path = "usa_state_abbreviations.csv"            # tabulka se zkratkami
key_main    = "State"                 # název sloupce se státem v Population_rase.csv
key_states  = "State"                 # název sloupce se státem ve states.csv (uprav podle reality)
value_col   = "Abbreviation"          # JEDINÝ sloupec, který chceš přidat (např. zkratka)


# 2) vyber jen klíč a cílový sloupec
st = states[[key_states, value_col]].copy()

# 3) znormalizuj klíče (odstranění mezer, sjednocení velikosti písmen)
df["__k"] = df[key_main].astype(str).str.strip().str.casefold()
st["__k"] = st[key_states].astype(str).str.strip().str.casefold()

# 4) ošetři duplicity ve states (vezmi první výskyt)
st = st.drop_duplicates("__k", keep="first")

# 5) připoj JEDINÝ sloupec do hlavní tabulky
df = df.merge(st[["__k", value_col]], on="__k", how="left").drop(columns="__k")

# 6) případně přejmenuj nový sloupec
df = df.rename(columns={value_col: "StateCode"})

# 7) ulož (volitelné)
df.to_csv("Population_rase_with_codes.csv", sep=";", index=False, encoding="utf-8-sig")
```

Prerazeni slupce StateCode na jine misto

```python
# Najdi pozici sloupce "State"
pos = df.columns.get_loc("State")

# Přesuň "StateCode" na pozici před "State"
cols = df.columns.tolist()
cols.insert(pos, cols.pop(cols.index("StateCode")))
df = df[cols]
```

```python
df.to_csv("Population_rase_with_codes.csv", sep=";", index=False, encoding="utf-8-sig")
```
