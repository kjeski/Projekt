```python
import pandas as pd

df = pd.read_csv("Population_rase_with_codes.csv", sep=";", encoding="utf-8-sig")
df.head()
```

```python
df_us = df[df["State"] == "U.S."]       
df_other = df[df["State"] != "U.S."]    


df_us.to_csv("Population_rase_US.csv", sep=";", index=False, encoding="utf-8-sig")
df_other.to_csv("Population_rase_other.csv", sep=";", index=False, encoding="utf-8-sig")
```
