```python
import pandas as pd


df = pd.read_csv("Population_Rase_final.csv", sep=";", encoding="utf-8-sig")
df.head()

```

```python
split_col = "White Alone"
split_index = df.columns.get_loc(split_col)

df1 = df.iloc[:, :split_index]  
df2 = pd.concat([df.iloc[:, :3], df.iloc[:, split_index:]], axis=1)

df1.to_csv("Population_part1.csv", index=False)
df2.to_csv("Population_part2.csv", index=False)
```
