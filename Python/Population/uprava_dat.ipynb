Smazani sloupcu

```python
import pandas as pd
df = pd.read_csv("Population_rase_other.csv", sep=";", encoding="utf-8-sig")
df.head()
```

```python
df = df.drop(columns=['Countyfips'])
df.head()
```

```python
df.to_csv("Population_rase_other.csv", sep=';', index=False)
```

Odstraneni 'Alone'

```python
df["Race"] = df["Race"].str.replace("Alone", "", regex=False).str.strip()
df.head()
```

```python
df.to_csv("Population_rase_other.csv", sep=';', index=False)
```

Smazani radku Not Hispanic

```python
df = df[df["Race"] != "Not Hispanic"]
```

```python
df.to_csv("Population_rase_other_woNotHispanic.csv", sep=';', index=False)
```

Prepsani ostatnich ras na Other

```python
df["Race"].unique()
```

```python
df["Race"] = df["Race"].replace(
    ["Hawaiian or Pacific Islander", "Two or More Races"],
    "Other"
)
```

```python
df.to_csv("Population_rase_other_woNotHispanic.csv", sep=';', index=False)
```

Zmena rasy Native Ameerican

```python
df["Race"] = df["Race"].replace(
    ["American Indian or Alaskan Native"],
    "Native American"
)
```

```python
df.to_csv("Population_rase_other_woNotHispanic.csv", sep=';', index=False)
```

Zmena rasy Hispanic

```python
df["Race"] = df["Race"].replace(
    ["Hispanic"],
    "Latino"
)
df.to_csv("Population_rase_other_woNotHispanic.csv", sep=';', index=False)
```

Zmena datoveho typu

```python
df = pd.read_csv("Population_rase_other_woNotHispanic.csv", sep=";", encoding="utf-8-sig")
df.dtypes
```

```python
df["Population"] = df["Population"].astype("Int64")
df.head()
```

```python
df.to_csv("Population_rase_other_woNotHispanic.csv", sep=';', index=False)
```
