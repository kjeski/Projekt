```python
import pandas as pd

df_main = pd.read_csv("States_status.csv", sep=";", encoding="utf-8-sig")
df_states = pd.read_csv("usa_state_abbreviations.csv", encoding="utf-8-sig")
```

```python
missing_in_states = df_main.loc[
    ~df_main["State"].isin(df_states["State"])
]["State"].unique()

print("Tyto státy jsou v States_statis, ale NEjsou v state_abbreviation:")
print(missing_in_states)
```
