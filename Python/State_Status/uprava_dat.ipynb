Zmena nazvu sloupce Status na Description



```python
import pandas as pd
df = pd.read_csv("States_status_with_codes.csv", sep=";", encoding="utf-8-sig")
df.head()
```

```python
df = df.rename(columns={"Status": "Description"})
df.head()
```

```python
df = df.drop(columns=["Last_Execution"])
df.head()
```

```python
df.head()
```

```python
df.loc[df["State"] == "District of Columbia", "Last_execution"] = 1957
df.head()
```

```python
df.to_csv("States_status_with_codes.csv", sep=";", index=False, encoding="utf-8-sig")
```

Kontrola datovych typu

```python
df.dtypes
```

```python
cols_to_convert = ["Year", "Last_execution"]

df[cols_to_convert] = df[cols_to_convert].astype("Int64")  
df.to_csv("States_status_with_codes.csv", sep=";", index=False, encoding="utf-8-sig")
```

Pridani sloupce Status

```python
df["Status"] = ""
```

Doplneni statusu pro staty


```python
df.loc[(2025 - df["Last_execution"]) > 10, "Status"] = "Moratorium"
```

```python
df.to_csv("States_status_with_codes.csv", sep=";", index=False, encoding="utf-8-sig")
```

```python
df.loc[df["Year"].notna(), "Status"] = "Abolition"
```

```python
df.to_csv("States_status_with_codes.csv", sep=";", index=False, encoding="utf-8-sig")
```

```python
df.loc[df["Description"].isna(), "Status"] = "Active"
df.to_csv("States_status_with_codes.csv", sep=";", index=False, encoding="utf-8-sig")
```

```python
df.dtypes
```

```python
df.loc[df["State"] == "District of Columbia", "Status"] = "Abolition"
df.to_csv("States_status_with_codes.csv", sep=";", index=False, encoding="utf-8-sig")
```

Serazeni statu podle nazvu podle abecedy


```python
df = pd.read_csv("States_status_with_codes.csv", sep=";", encoding="utf-8-sig")
df = df.sort_values(by="State", ascending=True)
df.head()
```

```python
df.to_csv("States_status_with_codes.csv", sep=";", index=False, encoding="utf-8-sig")
```

```python
position = df.columns.get_loc("State")

# Přesuň "StateCode" na pozici před "State"
cols = df.columns.tolist()
cols.insert(position, cols.pop(cols.index("Status")))
df = df[cols]
df.head()
```

```python
df.to_csv("States_status_with_codes.csv", sep=";", index=False, encoding="utf-8-sig")
```
